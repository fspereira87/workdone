```python
import math
import matplotlib.pyplot as plt
from matplotlib.pyplot import imread
import scipy
import tensorflow as tf
from tensorflow.python.framework import ops
import tensorflow.keras.layers as tfl
```

```python
train = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
train.head()
```

```python
test = pd.read_csv ('/kaggle/input/digit-recognizer/test.csv')
test.head()
```

```python
y_train = train.iloc[: , 0]
x_train = train.iloc[: , 1:]

print('Y Train', y_train)
print('X Train', x_train)
```

```python
x_train = x_train/255
```

```python
x_train = x_train.values.reshape(-1,28,28,1)
```

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train, num_classes = 10)
```

```python
x_test = test
x_test = x_test/255
x_test = x_test.values.reshape(-1,28,28,1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size = 0.2, random_state=42)
```

```python

model = tf.keras.Sequential([
    tfl.Conv2D(filters=64, kernel_size=3, padding='same', activation='relu', input_shape=(28,28,1)),
    tfl.Conv2D(filters=64, kernel_size=3, padding='same', activation='relu'),
    tfl.Conv2D(filters=128, kernel_size=3, padding='same', activation='relu'),
    tfl.MaxPooling2D(pool_size=2),
    tfl.Conv2D(filters=128, kernel_size=3, padding='same', activation='relu'),
    tfl.Conv2D(filters=192, kernel_size=3, padding='same', activation='relu'),
    tfl.MaxPooling2D(pool_size=2),
    tfl.Conv2D(filters=192, kernel_size=5, padding='same', activation='relu'),
    tfl.MaxPooling2D(pool_size=2, padding='same'),
    tfl.Flatten(),
    tfl.Dense(256, activation='relu'),
    tfl.Dense(10, activation='softmax')
])
model.compile(optimizer='adam',loss='categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_train, y_train, epochs = 10, batch_size=32)
```

```python
predictions = model.predict(x_test)
predict_labels = predictions.argmax(axis=1)
print(predict_labels)
```

```python
submission = pd.DataFrame({"ImageID": range(1, len(predict_labels) +1), "Label": predict_labels})
submission.to_csv("submission.csv", index = False)
```
